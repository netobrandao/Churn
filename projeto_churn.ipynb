# Algoritimos de Machine Learning

## Tratamento dos dados

```python
#separando os dados em variaveis explicativas e variavel alvo

x = dados.drop(['id_cliente','churn'], axis = 1)
y = dados['churn']
```

```python
x
```

```python
y
```

```python
#Tratandos os dados

#Modelos de machine learning não são capazes de trabalhar com variaveis em formato de texto, sendo necessaria a conversão

#Importando bibliotecas
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

#Inicializando
one_hot = make_column_transformer((
    OneHotEncoder(drop = 'if_binary'),
    ['sexo_biologico','pais', 'tem_cartao_credito', 'membro_ativo']
),
    remainder = 'passthrough',
    sparse_threshold=0)
```

```python
#Criando dataset 'x' com os nomes das colunas
xnames = x.columns #Armazenando nomes das colunas de 'x'
x = one_hot.fit_transform(x) #transformando dataset 'x'
x_transform = pd.DataFrame(x, columns = one_hot.get_feature_names_out(xnames))
x_transform
```

```python
#Transformando o dataset 'y'

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

y = label_encoder.fit_transform(y)
```

```python
#Datasets transformados
print(x,y)
```

```python
#Separação dos datasets de treino e teste

#Importando função
from sklearn.model_selection import train_test_split

#Aplicando função
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, stratify = y,  random_state = 42)
```

## Aplicando modelos

### DummyClassifier

```python
from sklearn.dummy import DummyClassifier #importando modelo

dummy = DummyClassifier(random_state=42) #instanciando o modelo
dummy.fit(x_treino, y_treino) #treinando o modelo

score_dummy = dummy.score(x_teste, y_teste) #score do modelo

print(f'A taxa de acerto do modelo DummyClassifier é de {score_dummy:.2%}')
```

```python
#Prevendo valores para o x_teste
y_preditodummy = dummy.predict(x_teste)
y_preditodummy 
```

```python
#Matriz de confusão
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

visualizacaotree = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_teste,y_preditodummy))
visualizacaotree.plot();
```

### DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier #importando modelo

tree = DecisionTreeClassifier(max_depth = 4, random_state =5) #instanciando o modelo

tree.fit(x_treino,y_treino) #treinando o modelo

score_dtc = tree.score(x_teste,y_teste) #score do modelo

print(f'A taxa de acerto do modelo DecisionTreeClassifier é de {score_dtc:.2%}')
```

```python
#Grafico DecisionTreeClassifier
from sklearn.tree import plot_tree

plt.figure(figsize = (15, 6)) #definindo o tamanho da figura
plot_tree(tree, filled = True, class_names = ['nao', 'sim'], fontsize = 5);
```

```python
#Prevendo resultados para o x_teste
y_preditotree = tree.predict(x_teste)
y_preditotree
```

```python
#Metricas DecisionTreeClassifier

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

print(f'Acuracia: {accuracy_score(y_teste,y_preditotree):.2%}')
print(f'Precisão: {precision_score(y_teste,y_preditotree):.2%}')
print(f'Recall: {recall_score(y_teste,y_preditotree):.2%}')
print(f'F1 Score: {f1_score(y_teste,y_preditotree):.2%}')
```

```python
#Matriz de confusão DecisionTreeClassifier

from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

visualizacaotree = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_teste,y_preditotree))
visualizacaotree.plot();
```

### KNN(KNeighborsClassifier)

```python
#KNeighborsClassifier

#Tratando dados para o modelo knn
from sklearn.preprocessing import MinMaxScaler

#Normalizando os dados
normalizacao = MinMaxScaler()
x_treino_normalizado = normalizacao.fit_transform(x_treino)
pd.DataFrame(x_treino_normalizado)
```

```python
#Importando modelo
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier() #instanciando modelo
knn.fit(x_treino_normalizado, y_treino) #treinando modelo
```

```python
#Normalizando x_teste
x_teste_normalizado = normalizacao.transform(x_teste)
```

```python
#Score do modelo
knn.score(x_teste_normalizado, y_teste)
```

```python
#Prevendo resultados para o x_teste_normalizado
y_preditoknn = knn.predict(x_teste_normalizado)
y_preditoknn
```

```python
#Metricas KNN
print(f'Acuracia: {accuracy_score(y_teste,y_preditoknn):.2%}')
print(f'Precisão: {precision_score(y_teste,y_preditoknn):.2%}')
print(f'Recall: {recall_score(y_teste,y_preditoknn):.2%}')
print(f'F1 Score: {f1_score(y_teste,y_preditoknn):.2%}')
```

```python
#Matriz de confusão para o knn
vizualizacaoknn = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_teste,y_preditoknn))
vizualizacaoknn.plot();
```
